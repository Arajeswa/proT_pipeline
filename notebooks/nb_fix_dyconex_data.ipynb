```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv("../data/input/Prozessdaten_MSEI_01_01_2022-07_07_2025_csv/plasma.csv", low_memory=False)

machines = df["Machine"].unique()
missing_val_cols = {machine: [] for machine in machines}
failed_mean_cols = {machine: [] for machine in machines}
for machine in tqdm(machines):
    df_ = df[df["Machine"] == machine].copy()
    for col in df_.columns:
        try:
            col_mean = df_[col].mean()
            if col_mean == 0:
                missing_val_cols[machine].append(col)
        except:
            failed_mean_cols[machine].append(col)
            continue
        
for machine in machines:
    df.loc[df['Machine'] == machine, missing_val_cols[machine]] = df.loc[df['Machine'] == machine, missing_val_cols[machine]].replace(0, np.nan)


# assert df.loc[df['Machine'] == machines[0], missing_val_cols[machine[0]]].isna().all(), "Missing values not replaced correctly for machine {}".format(machines[0])

df.to_csv("../data/input/Prozessdaten_MSEI_01_01_2022-07_07_2025_csv/plasma_fixed.csv", index=False)
```
