```python
import pandas as pd
import os

def convert_xlsx_to_csv(xlsx_path: str, output_dir: str = None, sheet_name: str = None) -> str:
    """
    Converts an Excel (.xlsx) file to a CSV file for faster reading in pandas.

    Args:
        xlsx_path (str): Path to the .xlsx file.
        output_dir (str): Directory to save the .csv file. Defaults to same directory as xlsx_path.
        sheet_name (str): Sheet name to convert. Defaults to the first sheet.

    Returns:
        str: Path to the generated .csv file.
    """
    # Use same folder if output_dir not specified
    if output_dir is None:
        output_dir = os.path.dirname(xlsx_path)

    # Get base file name without extension
    base_name = os.path.splitext(os.path.basename(xlsx_path))[0]
    csv_path = os.path.join(output_dir, base_name + ".csv")

    # Read and convert
    df = pd.read_excel(xlsx_path, sheet_name=sheet_name)
    df.to_csv(csv_path, index=False)

    return csv_path
```
